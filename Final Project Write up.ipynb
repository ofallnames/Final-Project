Thomas Stanton
Dataset: https://www.kaggle.com/competitions/boston-housing/overview
May 2, 2024
	
	My challenge for this project was to predict the value of houses in the Boston area near the Charles River. I first evaluated the data with no feature transformations through a neural network, random forest, multiple linear regression model, and a linear Lasso regression model. Then in order to evaluate the effect of race on the model, I removed the attribute black. This metric had already been transformed from the direct observation of the proportion of black people in the area. It is calculated as “1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town”. As you can see, this metric simply measures the proportion of black people different from 63%. There was no information available for where this number came from. I theorized it might have something to do with the average population of black people but was unable to find any numbers near that and discovered that it is based on the average percentage of white people in Boston around the 1980’s study this dataset was taken from. With my background in mathematics and limit theory, I was able to calculate that the limit of this metric was 396.9 when there were zero black people present in the population. And finally I ran these attributes through a principal component analysis, selected the top components and passed them through the same models. 


When I first analyzed the data I was left with relatively skewed distribution but luckily I had did not have any missing attribute for any of the recorded entries. Upon further analysis of the data for distribution and outliers it was obvious due to the large number of outliers on the bottom of the black metric and its skewed right distribution it is obvious that the majority of the houses surveyed in this data set was sampled from a majority of almost completely white neighborhoods. This could also explain a similar number of outliers in the crime metric. Unfortunately, due to the history of America’s racism, it can be inferred this might be affected by the policing in these areas and definitely has an affect on our study that is hard to capitulate in a normal data analysis. The metric chars also potentially affected the accuracy of our results during the first analysis because there is only 20 samples representative for all properties located on the Charles River. Also take careful note that this study is mostly applicable to houses with a number of bedrooms between 4-7 (single family homes). We had a few outliers above and below this metric, but most of the data was centered in this range. 
	After getting familiar with this data I ran it though a neural network, random forest, multiple linear regression model, and a Lasso Regression with the untransformed features with the corresponding RMSE on a testing set can be seen in Table 1. As per my expectations the multiple linear regression model performed the worst on the test set. This was expected as generally real estate prices work upon a critical mass spectrum, where a combination of linear factors end up creating exponential effect on not only the target variable but also the features used for training the model for prediction. This model would be useful to watch if applied to a certain area because once the prices derivate from the model, it could be a potential indicator of the area reaching said critical mass. Surprisingly, the neural network in this trial performed remarkably poorly. I altered the number of layers, number of nodes in each layer, and activation function and ended up settling on ReLU with two hidden layers of 20 and 10 nodes respectively. Any added layer significantly increased the RMSE while increasing complexity as well. I think this was due to the limited number of objects to study. We only had slightly over 300 data points and this clearly pales in comparison to the correct amount of data required to train a complex neural network on. The random forest produced the best results. This might be due to the bias introduced by the black attribute, as you will see in the next analysis this advantage is removed from the random forest model and it performs about as accurately as the other models. In order to, improve the Lasso regression. I attempted to run it through a 10-fold cross validation, but had problems with implementing the tensorFlow package from python in R code. I had to download the package and reforce instillation of tensorFlow in R studio every time I ran it with console commands in my python environment so I decided to exclude it from this project as it will not be able to recreated on your machine so I have stuck with comparing my four models performance based on pure attributes, pure attributes – black, and the PCA.


My analysis on the models without the black feature will be short. Removal of this attribute allowed the models to perform more similar as a group. However, it did affect the accuracy of the predictions of almost all models and make them perform worse overall. The only model to avoid any significant changes in the metric of their cost function was the neural network. This is probably due to the neural networks ability to handle heavy data transformations within its structure and filter out some of the noise using the ReLU function without losing the linearity of the price prediction.

After this I decided it was time to put these features through a principal component analysis to isolate interdimensionality of the features to a basis of an axis in a multiple dimensional space. Upon further analysis of the principal components, I decided to keep only 7 components which explained 90% of the variance while still eliminating 7 other. This did little to actually improve the accuracy of the MLM, LR, and RF model. This is indicative of the another underlying correlation I was unable to uncover. The fact that it slightly improved the MLM and LR model means there was collinearity present that the removal of helped eliminate the variance. The neural network’s accuracy improved significantly due to this transformation though. This is probably due to the fact that hand the neural network only a 7 dimensional space to “think about” with slightly over 300 observations is easier to come up with an explanation or system then a much more complex 14 dimensional space.

Much has changed in the housing market in the last 40 years.
